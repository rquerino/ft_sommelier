```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rw_data = pd.read_csv('resources/winequality-red.csv', delimiter=';')

# Function to map the colors as a list from the input list of x variables
def pltcolor(lst, good):
    cols=[]
    for l in lst:
        if l >= good:
            cols.append('green')
        else:
            cols.append('pink')
    return cols

x,y = rw_data.shape
fig, ax = plt.subplots(y, y, figsize=(70,70))
for i in range(0, y):
    for j in range(0, y):
        if (i != j):
            # Maybe the color has to be base only in the quality column?
            cols = pltcolor(rw_data.iloc[:, y - 1], 5)
            ax[i, j].scatter(rw_data.iloc[:, i], rw_data.iloc[:, j], c=cols)
        else:
            ax[i, j].text(0.5, 0.5, s=rw_data.iloc[:, i].name, fontsize=20, horizontalalignment='center', verticalalignment='center')
# def plot_scatter_matrix(wine_data, good_threshold, bad_threshold, save_plot=False):
# fig will be used to save the image
plt.show()
```
